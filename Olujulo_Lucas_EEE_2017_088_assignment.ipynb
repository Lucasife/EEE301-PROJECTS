EE301 Graph on Frequency(GHz) over Voltage(V)

```python
import numpy as np
import pandas as pd
from numpy import *
import matplotlib.pyplot as plt

freq = array([1869.18, 1672.99, 1446.83, 1181.3, 834.77, 1.000])
voltage = array([-0.110, -0.088, -0.066, -0.044, -0.022, 0])

fig, ax = plt.subplots()

plt.grid()

plt.title('Frequency(GHz) on Voltage(V).',fontsize=20, color = "blue")
plt.yticks(color='red')
plt.xticks(color='red')

plt.xlabel('Voltage(V)', fontsize=12)
plt.ylabel('Frequency(GHz)',fontsize=12)

line, = ax.plot(voltage,freq, color="green")
f = np.linspace(0,10,100)

plt.scatter(voltage,freq, color ='red')

plt.show()
```
